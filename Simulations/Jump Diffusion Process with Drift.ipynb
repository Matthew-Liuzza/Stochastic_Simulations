```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from numpy import random as rn
from scipy import stats
import scipy.stats as si

G = 1629562571
np.random.seed(G)
from datetime import datetime
now = datetime.now()
dt_string = now.strftime("%d/%m/%Y %H:%M:%S")

S0 = 1
T = 10.0
mu = 0.05
sigma = 0.25
lam = 0.05
theta = 1.1
M = 50000
N = 100

epsilon = rn.randn(M,N)
F_lam = rn.poisson(lam,(M,N))
S = S0*np.ones((M,N+1))
dt = T/N

start_time = datetime.now()
for i in range(0,N):
    S[:,i+1] = S[:,i] * (1 + theta * F_lam[:,i] * dt + mu * dt + sigma * epsilon[:,i] * np.sqrt(dt))
    
plt.figure(figsize=(13,7))
fontsize=15
plt.title('Path-Dependent Monte Carlo Simulation - Jump Diffusion Process with Drift',fontsize=fontsize)
plt.xlabel('Years',fontsize=fontsize)
plt.ylabel('FX prices (USD)',fontsize=fontsize)
plt.grid()
a = [ rn.randint(0,M) for j in range(1,8)]
for runer in a:
    plt.plot(np.arange(0,T+dt,dt), S[runer], 'red')
```
